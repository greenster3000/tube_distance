```python
import pandas as pd
import folium as f
```

```python
df = pd.read_csv("tube_lat_lon.csv")
```

```python
len(df)
```

```python
df.head(1)
```

```python
m = f.Map()
m.fit_bounds([[df.y.min(), df.x.min()],
              [df.y.max(), df.x.max()]])
for t in df.itertuples():
    f.Circle((t.y, t.x),
            radius=805,
            fill_color="black",
            color="black",
            popup=t.NAME,
            fill_opacity=1.0).add_to(m)
m
```
